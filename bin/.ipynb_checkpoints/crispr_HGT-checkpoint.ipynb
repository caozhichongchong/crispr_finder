```python
# read multiple seq alignment
windowsize = 10
dat=pd.read_csv('data/WGS/HGT_justcrispr/all.BL.widespreadspacer.neighbour.fasta.align.SNPcount.txt', sep='\t')
dat.loc[dat['SNPs']<0,'SNPs']=0
dat['SNPstrains'] = dat['SNPs']*dat['coverage']*(1-dat['MAF'])
dat['SNPs_loci']=dat['SNPs']
dat.loc[dat['SNPs_loci']>1,'SNPs_loci']=1
#dat['locusrange']=[int(x/windowsize) for x in dat['locus']]
dat.head()
```

```python
# moving window
mut_rate = 1/1000000 # 1 SNP per year per 1 mbp
mut_rate_1kb = mut_rate*1000
dat = dat[dat['coverage']>=20]
dat['SNPs_windows'] = 0
dat['SNPs_loci_windows']=0
dat.index = range(0,dat.shape[0])
for i in range(0,dat.shape[0]- windowsize):
        dat.loc[i,'SNPs_windows'] = dat.loc[i:(i+windowsize),'SNPs'].sum()
        dat.loc[i,'SNPs_loci_windows'] = dat.loc[i:(i+windowsize),'SNPs_loci'].sum()
        if i%10000 == 0:
            print('processed %s lines'%(i))
dat2 = dat
dat2['SNPs_loci_windows']=dat2['SNPs_loci_windows']/windowsize*1000 # average NO. SNPs per 1kb
dat2['SNPs_windows']=dat2['SNPs_windows']/windowsize*1000 # average NO. SNPs per 1kb
dat2['years_since'] = dat2['SNPs_windows']/mut_rate_1kb
dat2.to_csv('data/WGS/HGT_justcrispr/all.BL.widespreadspacer.neighbour.fasta.align.SNPcountsum.txt', sep='\t')
dat2.head()
```

```python
def find_loci(alignseq):
    newloci = dict()
    i = 0
    j = 0
    for seq in alignseq:
        j += 1
        if seq != '-':
            i += 1
            newloci.setdefault(i,j)
    return newloci
newloci = find_loci(alignseq)
newloci = pd.DataFrame.from_dict(newloci,orient = 'index',columns = ['newloci'])
newloci['oldloci']=newloci.index
newloci.head()
```

```python
# topology sorting
#Python program to print topological sorting of a DAG
from collections import defaultdict

#Class to represent a graph
class Graph:
	def __init__(self,vertices):
		self.graph = defaultdict(list) #dictionary containing adjacency List
		self.V = vertices #No. of vertices

	# function to add an edge to graph
	def addEdge(self,u,v):
		self.graph[u].append(v)

	# A recursive function used by topologicalSort
	def topologicalSortUtil(self,v,visited,stack):

		# Mark the current node as visited.
		visited[v] = True

		# Recur for all the vertices adjacent to this vertex
		for i in self.graph[v]:
			if visited[i] == False:
				self.topologicalSortUtil(i,visited,stack)

		# Push current vertex to stack which stores result
		stack.insert(0,v)

	# The function to do Topological Sort. It uses recursive
	# topologicalSortUtil()
	def topologicalSort(self):
		# Mark all the vertices as not visited
		visited = [False]*self.V
		stack =[]

		# Call the recursive helper function to store Topological
		# Sort starting from all vertices one by one
		for i in range(self.V):
			if visited[i] == False:
				self.topologicalSortUtil(i,visited,stack)

		# Print contents of stack
		return stack

g= Graph(len(allgeneclusterID))
for ref in AAHGTorder['ref'].unique():
    AAHGTordersub = AAHGTorder[AAHGTorder['ref']==ref]
    # check order
    AAHGTordersub = AAHGTordersub.merge(anno,left_on = 'AA1',right_on = 'gene_name',how = 'left')
    try:
        cas9 = int(AAHGTordersub.loc[AAHGTordersub['EGGNOG']=='COG3513','geneID'])
        cas1 = int(AAHGTordersub.loc[AAHGTordersub['EGGNOG']=='COG3512','geneID'])
        if cas9 > cas1:
            AAHGTordersub = AAHGTordersub.sort_values('geneID',ascending=False)
            print('reorder %s %s %s'%(ref,cas9,cas1))
            # need reorder
        for i in range(0,AAHGTordersub.shape[0]-1):
            g.addEdge(AAHGTordersub.iloc[i,4], AAHGTordersub.iloc[i+1,4])
    except TypeError:
        print(ref)

print ("Following is a Topological Sort of the given graph")
allorder = g.topologicalSort()
print(allorder[:10])
```
