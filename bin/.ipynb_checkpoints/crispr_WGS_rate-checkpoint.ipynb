```python
# MG spearman
SP_acquired_info= pd.read_csv('data/MG/all.spacers.acquired.txt', sep='\t')
SP_acquired_info['DR']=[x.split('_')[0] for x in SP_acquired_info['DR']]
SP_acquired_info['Acquired']=False
SP_acquired_info.loc[SP_acquired_info['qvalue']==0,'Acquired']=True
SP_acquired_info_sum = pd.DataFrame(SP_acquired_info[SP_acquired_info['Acquired']==True].groupby(['DR'])['Acquired'].count())
SP_acquired_info_sum.reset_index(inplace=True)
DR_abu_spearman = pd.read_csv('data/MG/DR_abun_sample.csv')
DR_abu_spearman['DR']=[x.upper() for x in DR_abu_spearman['DR']]
DR_abu_spearmansamplesum = list(DR_abu_spearman.sum())[1:]
for i in range(1,DR_abu_spearman.shape[1]):
    DR_abu_spearman.iloc[:,i] = DR_abu_spearman.iloc[:,i]/DR_abu_spearmansamplesum[i-1]
DR_abu_spearman['abu_mean'] = 0
for i in DR_abu_spearman.index:
    DR_abu_spearman.loc[i,'abu_mean'] = np.mean(list(DR_abu_spearman.loc[i,:])[1:DR_abu_spearman.shape[1]-1])
DR_abu_spearman = DR_abu_spearman.merge(SP_acquired_info_sum,left_on = 'DR',right_on = 'DR',how = 'left')
DR_abu_spearman.loc[DR_abu_spearman['Acquired'].isna(),'Acquired']=0
DR_abu_spearman = DR_abu_spearman.loc[:,['DR','abu_mean','Acquired']]
#DR_abu_spearman['abu_mean']=np.log10(DR_abu_spearman['abu_mean'])
DR_abu_spearman.head()
```

```python
# random forest importance example
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = boston.target
# split x,y, test,training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
rf.feature_importances_
```

```python
# random forest importance load data
# spearman test
DR_abu_spearman_all = pd.DataFrame()
reads_per_DR_set = [6,7,8,9,10,15,20]
for reads_per_DR in reads_per_DR_set:
    # read original data
    SP_acquired_info= pd.read_csv('data/MG/all.spacers.acquired.txt', sep='\t')
    SP_acquired_info['DR']=[x.split('_')[0] for x in SP_acquired_info['DR']]
    SP_acquired_info['Acquired']=False
    SP_acquired_info.loc[SP_acquired_info['qvalue']==0,'Acquired']=True
    SP_acquired_info_sum = pd.DataFrame(SP_acquired_info[SP_acquired_info['Acquired']==True].groupby(['DR'])['Acquired'].count())
    SP_acquired_info_sum.reset_index(inplace=True)
    DR_abu_spearman = pd.read_csv('data/MG/DR_abun_sample.csv')
    DR_abu_spearman['DR']=[x.upper() for x in DR_abu_spearman['DR']]
    DR_abu_spearmansamplesum = list(DR_abu_spearman.sum())[1:]
    for i in range(1,DR_abu_spearman.shape[1]):
        DR_abu_spearman.iloc[:,i] = DR_abu_spearman.iloc[:,i]/DR_abu_spearmansamplesum[i-1]
    DR_abu_spearman['abu_mean'] = 0
    for i in DR_abu_spearman.index:
        DR_abu_spearman.loc[i,'abu_mean'] = np.mean(list(DR_abu_spearman.loc[i,:])[1:DR_abu_spearman.shape[1]-1])
    DR_abu_spearman = DR_abu_spearman.merge(SP_acquired_info_sum,left_on = 'DR',right_on = 'DR',how = 'left')
    DR_abu_spearman.loc[DR_abu_spearman['Acquired'].isna(),'Acquired']=0
    DR_abu_spearman = DR_abu_spearman.loc[:,['DR','abu_mean','Acquired']]
    # MG spearman reads_set
    SP_acquired_info= pd.read_csv('data/MG/all.spacers.acquired_rarefy_%sreads.txt'%(reads_per_DR), sep='\t')
    SP_acquired_info['DR']=[x.split('_')[0] for x in SP_acquired_info['DR']]
    SP_acquired_info['Acquired']=False
    SP_acquired_info.loc[SP_acquired_info['qvalue']==0,'Acquired']=True
    SP_acquired_info_sum = pd.DataFrame(SP_acquired_info[SP_acquired_info['Acquired']==True].groupby(['DR'])['Acquired'].count())
    SP_acquired_info_sum.reset_index(inplace=True)
    DR_abu_spearman_subset = pd.read_csv('data/MG/DR_abun_sample.csv')
    DR_abu_spearman_subset['DR']=[x.upper() for x in DR_abu_spearman_subset['DR']]
    DR_abu_spearman_subsetsamplesum = list(DR_abu_spearman_subset.sum())[1:]
    for i in range(1,DR_abu_spearman_subset.shape[1]):
        DR_abu_spearman_subset.iloc[:,i] = DR_abu_spearman_subset.iloc[:,i]/DR_abu_spearman_subsetsamplesum[i-1]
    DR_abu_spearman_subset['abu_mean'] = 0
    for i in DR_abu_spearman_subset.index:
        DR_abu_spearman_subset.loc[i,'abu_mean'] = np.mean(list(DR_abu_spearman_subset.loc[i,:])[1:DR_abu_spearman_subset.shape[1]-1])
    DR_abu_spearman_subset = DR_abu_spearman_subset.merge(SP_acquired_info_sum,left_on = 'DR',right_on = 'DR',how = 'left')
    DR_abu_spearman_subset.loc[DR_abu_spearman_subset['Acquired'].isna(),'Acquired']=0
    DR_abu_spearman_subset = DR_abu_spearman_subset.loc[:,['DR','abu_mean','Acquired']]
    #DR_abu_spearman_subset['abu_mean']=np.log10(DR_abu_spearman_subset['abu_mean'])
    DR_abu_spearman = DR_abu_spearman.merge(
        DR_abu_spearman_subset.loc[:,['DR','Acquired']],left_on='DR',right_on='DR',how='left'
    )
    DR_abu_spearman['reads_depth']=reads_per_DR
    DR_abu_spearman_all = DR_abu_spearman_all.append(DR_abu_spearman)

DR_abu_spearman_all.to_csv('data/all.spacers.acquired.allreaddepth.txt', sep='\t',index=False)
DR_abu_spearman_all.tail()

```

```python
# random forest importance
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X = DR_abu_spearman_all.loc[:,['abu_mean','reads_depth']]
y = DR_abu_spearman_all['Acquired_y']
# split x,y, test,training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
print(rf.feature_importances_)
plt.barh(['abu_mean','reads_depth'], rf.feature_importances_)
plt.xlabel("Random Forest Feature Importance")
plt.savefig('data/MGSpearman_Depth_Abu_Predict_Acquisition_importance.pdf',bbox_inches='tight')
```

```python
import shap
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar")
plt.savefig('data/MGSpearman_Depth_Abu_Predict_Acquisition_importance2.pdf',bbox_inches='tight')
```

```python
# predict accuracy
from sklearn import metrics
y_pred_test = rf.predict(X_test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_test))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_test))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
#RMSE is the square root of MSE
```

```python
### infection rate for diff species
### Abu VS spacer acquisition rate
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series, DataFrame,date_range
from matplotlib import cm
import seaborn as sns
import scipy as sp
import math
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
dat=pd.read_csv('data/WGS/all.spacerchange.sum.gain.all.txt', sep='\t')
species_all = pd.read_csv('data/WGS/species_order.new.fullname.txt', sep='\t')
dat=dat[dat['No.SNP']>0]
dat=dat[dat['No.crispr']>0]
dat = dat.drop_duplicates(['CL'])
dat=dat.merge(species_all.loc[:,['species','fullname']],
             left_on = 'Species',right_on = 'species',how='left')
dat=dat.sort_values('fullname')
dat['SP gain'] = dat['total_gain'] + 0.5 # assuming lineages with no SP gain is half way towards gaining one
dat['SP gain rate (SP gain/No. SNPs/No.crispr)']=dat['SP gain']/dat['No.SNP']/dat['No.crispr']
dat['No. generations'] = 1/dat['SP gain rate (SP gain/No. SNPs/No.crispr)']*300
dat2 = pd.DataFrame(dat.groupby('Species')['No. generations'].median())
dat2.columns=['No. generations mean']
dat2.reset_index(inplace=True)
dat=dat.merge(dat2,left_on = 'Species',right_on='Species',how = 'left')
dat=dat.sort_values('No. generations mean')
#dat['Pin'] = [min(df['Pin'][df['y']<=int(x)]) for x in dat['No. generations']]
dat['fullname'] = [x.replace('_',' ').replace(' 1','') for x in dat['fullname']]
dat['species']=dat['Species']
dat2 = pd.DataFrame(dat.groupby('Species')['CL'].count())
dat2.columns=['No.lineage total']
dat2.reset_index(inplace=True)
dat = dat.merge(dat2,left_on = 'Species',right_on = 'Species')
#dat['No.genome total']=[int(x) for x in dat['No.genome total']]
dat['Species']=['%s (n=%0d)'%(dat.loc[i,'fullname'],int(dat.loc[i,'No.lineage total'])) for i in dat.index]
dat.head()
```

```python
from scipy.stats import poisson
dat['No. generations log']=np.log10(dat['No. generations'])
fig, axs = plt.subplots(figsize=(8, 6))
sns.histplot(data=dat,x='No. generations log',ax = axs)
axs.plot([1,2,3,4,5,6], poisson.pmf([1,2,3,4,5,6], 4)*dat.shape[0], 'bo-', ms=8)
```

```python
from numpy import median
fig, axs = plt.subplots(nrows=1,ncols = 1,figsize=(12, 6),sharex='col')
sns.pointplot(data=dat, x = 'Species',y = 'No. generations',hue = 'Species',palette='rocket',
              estimator=median,ci=None,ax = axs)
sns.boxplot(data=dat, x = 'Species',y = 'No. generations',hue = 'Species',palette='rocket',dodge=False,
            ax = axs)
axs.set(yscale="log")
plt.sca(axs)
plt.xticks(rotation=90,style='italic')
#plt.yticks(ticks = [30,365/2,365,3650/2,3650,36500/2,36500,365000],labels = ['1 month','6 months','1 year','5 years','10 years','50 years','100 years','1000 years'])
plt.legend(bbox_to_anchor=(1.0, 1.0))
axs.set_xlabel('')  
axs.set_ylabel('No. generations to acquire a spacer') 
axs.get_legend().remove()
plt.savefig('data/WGS//CRISPR_rate_species.png',bbox_inches='tight',dpi = 300)
```
